```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
tou = 0.5

x_train = np.array(list(range(3, 33)))
print(x_train)
x_train = x_train[: , np.newaxis]
print(x_train)
y_train = np.array([1,2,1,2,1,1,3,4,5,4,5,6,5,6,7,8,9,10,11,11,12,11,11,10,12,11,11,10,9,8])
print(y_train)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
#testset train
x_test = np.array([i/10. for i in range(400)])
print(x_test)
x_test = x_test[:, np.newaxis]
print(x_test)
y_test = []
print(y_test)
```

```python
x_test.shape
```

```python
for r in range(len(x_test)):
    wts = np.exp(-np.sum((x_train-x_test[r])**2, axis=1))/(tou*2)**2
    W = np.diag(wts)
    factor1 = np.linalg.inv((x_train.T).dot(W).dot(x_train))
    parameters = factor1.dot((x_train.T).dot(W).dot(y_train))
    prediction = x_test[r].dot(parameters)
    
    y_test.append(prediction)
```

```python
y_test = np.array(y_test)
plt.scatter(x_train, y_train)
plt.scatter(x_test, y_test)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
